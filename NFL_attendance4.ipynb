```python
import requests
from bs4 import BeautifulSoup
import urllib.parse
import pymongo
import pandas as pd
from html_table_parser import HTMLTableParser
from pprint import pprint
import bs4
```

```python
# nfl url
url = f"https://www.sportsmediawatch.com/nfl-tv-ratings-viewership-nbc-cbs-fox-espn-nfln-regular-season-playoffs/"
print(url)
```

```python
response = requests.get(url)
print(response.status_code)
```

```python
#soup = BeautifulSoup(response.content, 'html')
#for trow in soup.select("table"):
#    data = [[' '.join(item.text.split()) for item in tcel.select("td")]
#            for tcel in trow.select("tr")]
#    print(data)
soup = bs4.BeautifulSoup(response.content, 'html.parser')
print(soup)
```

```python
results = soup.find(id='olytable')
print(results.prettify())
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
attendance = soup.find_all(id = 'olytable')
print(attendance)
```

```python
rows = soup.find_all('tr')
print(rows[:200])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(200)
```

```python
df1 = df[0].str.split(', ', expand=True)
df1.head(200)
```

```python
client = pymongo.MongoClient('mongodb://localhost:27017')
db=client.NFL_viewership
#col_reviews = db.tv_viewership2019
#col_index = db.tv_viewership2019
db.insert_many(df1.to_dict('records'))
#print(reviews)
```
